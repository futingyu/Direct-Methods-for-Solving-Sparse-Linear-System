# Abstract:
<p style='text-align: justify;'> Multiple methods are available in the literature to solve a linear system. They are generally classified in two categories, direct methods and iterative methods. In this work, we focus on solving a sparse linear system $ Ax = b$, where $A$ is symmetric positive definite. In this case, forward and backward triangular solve can be used after the Cholesky decomposition of matrix $A$ $\left( A = LL^T\right)$ has been computed. In this work, we explore the use of Block Low- Rank Compression(BLR) in the left-looking Cholesky algorithm. To this end, we investigate sorting the updaters to a certain target block during the factorization process and it's impact on rank growth.we compare the results with a regular factorization process without sorting the updates to see the growth of the rank of the target block. </p>

# Direct Methods for Solving Sparse Linear System
<p style='text-align: justify;'>
A sparse linear system is written in the form :
\begin{equation} \label{ eq:1} \tag{1}
Ax = b,
\end{equation}


where $A \in \mathbb{R}^{n \times n}$ is a sparse matrix , $x \in \mathbb{R}^n $ is the unknown vector and $b \in \mathbb{R}^n$ is the right-hand side vector. Generally, two types of methods are commonly used to solve the system ([1](#mjx-eqn-eq:1)). _Iterative methods_ build the sequence of iteration converging to the exact solution. _Direct methods_ build a factorization of A. For example,factor $A$ as $A = LU$ or $A = QR$ then solve the problem using the factorized form of A. In this work, we discuss a direct method. Also, we consider the matrix $A$  to be _Positive definite_ , meaning $A$ is real symmetric and $ x^TAx >0 , \  \forall \  x\neq 0$. In this situation, we factorize a matrix $A$ into triangular factorization,
\begin{equation} \label {eq:2} \tag{2}
A = LL^T
\end{equation}

  which is called _Cholesky factorization_ of a matrix $A$. The _Cholesky factor  L_ is a lower triangular matrix with positive diagonal elements. Using equations ([1](#mjx-eqn-eq:1)) and ([2](#mjx-eqn-eq:2)) yields
  \begin{equation*}
  LL^Tx = b,
  \end{equation*}
  and we substitute $y = L^Tx$ which provides the value of $x$ by solving the following triangular systems
  \begin{equation*}
  Ly = b
  \end{equation*}
  and
  \begin{equation*}
  L^Tx = y.
  \end{equation*}
